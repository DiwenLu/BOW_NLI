```python
import pandas as pd
import pickle as pkl


best_log = pkl.load(open("BEST_LOGISTIC_10000_50_hadamard.p", "rb"))
best_nn = pkl.load(open("BEST_NN_50000_50_concat.p", "rb"))
```

```python
max(best_nn['per_batch']['val_accs'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = np.arange(0, len(best_log['per_batch']['train_accs']))
train_acc = best_log['per_batch']['train_accs']
val_acc = best_log['per_batch']['val_accs']
train_loss = best_log['per_batch']['train_losses']
val_loss = best_log['per_batch']['val_losses']

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('iterations')
ax1.set_ylabel('acc', color=color)
ax1.plot(t, train_acc, color=color, linewidth=1, label = 'train_acc')
ax1.plot(t, val_acc, color=color, linestyle='dashed', linewidth=1, label = 'val_acc')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('loss', color=color)  # we already handled the x-label with ax1
ax2.plot(t, train_loss, color=color, linewidth=1, label = 'train_loss')
ax2.plot(t, val_loss, color=color, linestyle='dashed', linewidth=1, label = 'val_loss')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='lower right')
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('best LR (v_size = 10k, embed_dim = 50, hadamard)')
plt.show()
```
